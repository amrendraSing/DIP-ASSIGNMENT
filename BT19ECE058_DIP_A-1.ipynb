```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("amar.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
```

# Task 1

```python
gray = 0.2989*img[:,:,0] + 0.5870*img[:,:,1] + 0.1140*img[:,:,2]
```

```python
plt.imshow(gray, cmap='gray')
```

# Task2

```python
threshold = 127
bin = np.zeros((img.shape[0], img.shape[1]))
```

```python
for i in range(len(gray)):
    for j in range(len(gray[i])):
        if(gray[i,j] > threshold):
            bin[i,j] = 1
```

```python
plt.imshow(bin, cmap='binary')
```

# Task 3

```python
bin_gray = bin + gray
```

```python
plt.imshow(bin_gray, cmap='gray')
```

# Task 4

```python
gray_20 = gray + 20
```

```python
plt.imshow(gray_20, cmap='gray')
```
